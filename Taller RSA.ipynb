<p style='text-align: justify;'>
    <b>Problema 1:</b>
       Se debe especificar cuál es el problema que resuelve RSA, y por qué es    necesario introducir un algoritmo de este tipo. 
    </p>
<b>Solución</b>
    <p style='text-align: justify;'>
        Imaginemos que queremos enviar información muy importante a un destinatario. Una solución podría ser la de introducir el mensaje en una caja fuerte que tuviera una determinada combinación para poder abrirla. Luego, enviamos la caja al destinatario y este podrá abrirla únicamente si le proporcionamos la combinación (a esto se le conoce como criptografía simétrica). 
    <br>
        No obstante, el problema en este caso sería el intercambio de la clave ya que debemos proporcionársela al destinatario y en este camino podría ser interceptada.
    <br>
        Por otro lado, si el destinatario tuviera su propia caja fuerte con dos claves, una para abrir y otra para cerrar, entonces este podría recibir mensajes enviando la caja abierta y la clave para cerrarla al emisor de modo que este último pueda introducir el mensaje y cerrar la caja con la clave para cerrar. De esta manera, se solucionaría el problema de que la clave para abrir pueda ser interceptada durante el camino ya que ésta nunca se envió. A esto se le conoce como criptogría asimétrica, y el algortimo RSA es uno de los más utilizados actualmente ya que elimina el problema del intecambio de la clave.
    </p>

        <li>
            Computamos $\phi(n)$ donde $\phi$ es las función <i>totient</i> de Euler que se definde como:
            <br>
            <br>
            <b>Definición:</b> La función $phi$ de Euler o <i>toitent</i> de Euler es:
            <br>
            <br>
            $$
                \phi(n)=|\left\lbrace x: 1\leq x \leq n \hspace{0.2cm} y \hspace{0.2cm} mcd(x,n)=1\right\rbrace|
            $$
            <br>
            el número de todos los enteros positivos $x\leq n$ que no tienen divisores comunes con $n$.
            <br>
            <br>
            Note que si $n$ es primo, entonces $\phi(n)= n-1$ debido a que $mcd(x,n)=1$ a menos que $x$ divida a $n$.
            <br>
            Antes de profundizar en este ítem mostraremos el siguiente corolario y lema.
            <br>
            <br>
            <b>Corolario:</b> Si $m=nq+r$, entonces $mcd(m,n)=mcd(r,n)$.
            <br>
            <br>
            <b>Demostración:</b>
            <br>
            <br>
            Supongamos que $mcd(m,n)=d$, entonces tendríamos que $m=nq+r=dk_{1}$ y $n=dk_{2}$. Así, despejando de la primera ecuación tendríamos que 
            $$r=dk_{1}-nq=dk_{1}-(dk_{2})q=d(k_{1}-k_{2}q)$$ 
            y entonces vemos que $d$ también divide a $r$ de modo que $mcd(m,n)=mcd(r,n)$.
            <br>
            <br>
            <b>Lema:</b> Si $mcd(m,n)=1$, entonces $\phi(n*m)=\phi(n)*\phi(m)$.
            <br>
            <br>
            <b>Demostración:</b>
            <br>
            <br>
            Consideremos primero que: 
            <br>
            <br>
            $$
            \phi(nm)= |\left\lbrace x: 1\leq x \leq nm \hspace{0.2cm} y \hspace{0.2cm} mcd(x,nm)=1\right\rbrace|
            $$
            <br>
            <br>
            Por otro lado, coloquemos los números desde 1 hasta $nm$ en filas y columnas de la siguiente manera:

            <br>
            <br>
            Ahora lo que nos interesa saber es cuántos números de la tabla anterior son coprimos con $n$ y con $m$ ya que si son coprimos con cada uno entonces son coprimos con el producto. Sabemos que cada número de la tabla es de la forma $qm+r$ con $0\leq q \leq(n-1)$ y $1\leq r\leq m$ y también que por el corolario anterior tendríamos que $mcd(qm+r,m)=mcd(r,m)$. Por ejemplo, si cogemos la columna 3 tendríamos que $mcd(qm+3,m)=mcd(3,m)$ y esto suguiere que si $mcd(3,m)=1$ entonces todos los números de esa columna serían primos relativos con $m$ y como lo que nos interesa saber es cuántos primos relativos hay con $m$ entonces bastaría saber el número de columnas con primos relativos con $m$ y este número correspondería a $\phi(m)$. De manera similar ahora nos podríamos preguntar dentro de cada columna cuántos primos relativos hay con $n$, si consideramos una columna $r$ (números de la forma $qm+r$) y suponemos que ésta se compone de primos relativos con $m$ entonces tendríamos que si $mcd(r,n)=1$, entonces $mcd(qm+r,n)=1$, si mcd(qm,n)=1. Por hipótesis como $mcd(m,n)=1$ y $0\leq q \leq(n-1)$,si $mcd(q,n)=1$, entonces se cumpliría que $mcd(qm+r,n)=1$ y como hay $n-1$ números en cada columna tendríamos que existirían $\phi(n)$ números que son coprimos con $n$. De esta manera tendríamos que e cada columna habrían $\phi(n)$ primos relativos con $n$ y como hay $\phi(m)$ columnas con primos relativos con $m$ tendríamos que habrían $\phi(nm)=\phi(n)*\phi(m)$ primos relativos con $nm$.
            <br>
            <br>
            Teniendo en cuenta lo anterior, entonces $\phi(n)=\phi(p*q)=\phi(p)*\phi(q)=(p-1)(q-1)$ debido a que $p$ y $q$ son primos y $mcd(p,q)=1$ por la misma razón. Mantenemos $\phi(n)$ en secreto ya que se utilizará más adelante para calcular la llave de decriptación $d$.
        </li>
        <br><br>
        <li> Escogemos un $e<\phi(n)$ de modo que $e$ y $\phi(n)$ sean coprimos, es decir, que $mcd(e,\phi(n))=1$. De esta manera, $e$ será nuestra llave de encriptación o llave pública que le servirá al emisor para encriptar el mensaje.
            <br>
            Para determinar $e$ se escogerá un número aleatorio y se comprobará que este sea coprimo con $\phi(n)$ mediante el Algoritmo de Euclides, el cual se enuncia a continuación:
            <br>
            <br>
            <div style="border: ridge #000000 1px; padding: 10px 10px 10px 10px;">
                <b>Entrada:</b> Dos números enteros $a$ y $b$.
                    <br>
                <b>Salida:</b> El máximo común divisor entre $a$ y $b$.
                <ol>
                    <li>$r_{0}\leftarrow a$,$r_{1}\leftarrow b$.</li>
                    <li>$i\leftarrow 1$.</li>
                    <li>Mientras $r_{i}\neq 0$:
                        <ol>
                            <li>$r_{i+1}=r_{i-1}\hspace{0.1cm} mod \hspace{0.1cm} r_{i}$</li>
                            <li>$i=i+1$</li>
                        </ol>
                    </li>
                    <li>El resultado es: $r_{i-1}$.</li>
                </ol>
            </div>
            <br>
            <br>
            De esta manera, si $r_{i-1}=1$ entonces quiere decir que $e$ y $\phi(n)$ son coprimos.
            <br>
            <br>
        </li>
        <li>
            Ahora procederemos a calcular la llave de decriptación la cual es la llave que se mantendrá en secreto. Se tiene que $d$ satisface que $d*e\equiv 1\hspace{0.1cm} mod\hspace{0.1cm} \phi(n)$ o de la misma manera $d\equiv e^{-1}\hspace{0.1cm} mod\hspace{0.1cm} \phi(n)$ lo cual quiere decir que $d$ es el inverso multiplicativo de $e$ modulo $\phi(n)$ y debe existir debido a que $mcd(e,\phi(n))=1$ ya que por el Algoritmo extendido de Euclides $1=\alpha*e+\beta*\phi(n)$ y despejando tendríamos que $1-\beta*\phi(n)=\alpha*e$ lo cual suguiere que $\alpha*e\cong1\hspace{0.1cm}mod\hspace{0.1cm}\phi(n)$ para algún $\alpha \in \mathbb{Z}$.
            <br>
            Podemos calcular $d$ mediante el algoritmo extendido de Euclides el cual se enuncia a continuación:
            <br>
            <br>
            <div style="border: ridge #000000 1px; padding: 10px 10px 10px 10px;">
                <b>Entrada:</b> Dos números enteros $a$ y $b$.
                    <br>
                <b>Salida:</b> El inverso modular de $b$.
                <br>
                $(s,t,s\prime,t\prime)\leftarrow (1,0,0,1)$
                <br>
                <b>Mientras</b> $b\neq 0$ <b>hacer lo siguiente: </b>
                <blockquote>Divida $a$ entre $b$ para obtener un cociente $q$ y un residuo $r$
                <br>
                $(a,s,t,b,s\prime,t\prime)\leftarrow(b,s\prime,t\prime,r,s-s\prime q,t-t\prime q)$
                </blockquote>
                <br>
                <b>El resultado es $t$</b>
            </div>
        </li>
    </ol>
    <br>
        <br>
        <b>Distribución de las llaves.</b>
        <br>
        <br>
        La clave pública será $(e,n)$ la cual será enviada al emisor para que este encripte el mensaje. Por otro lado, la clave privada será $(d,n)$ con la cual el receptor decripatará el mensaje que le fue enviado. 
        <br>
        <br>
        <b>Encriptación (o cifrado).</b>
        <br>
        <br>
        Para que el emisor pueda cifrar el mensaje $M$ entonces tendrá que convetirlo en un número entero $m<n$ mediante un protocolo reversible acordado con anterioridad. Con esto, el emisor computa el texto cifrado $c$ usando la llave púeblica del receptor de la siguiente forma: $m^{e}\cong c\hspace{0.1cm}mod\hspace{0.1cm}n$. Esto puede hacerse rápido utilizando exponenciacón binaria la cual nos permite elevar números a grandes potencias de manera eficiente y gracias a las propiedades de la exponenciación modular podemos obtener el mensaje cifrado de manera rápida como se verá en la implementación del algoritmo.
        <br>
        <br>
        <b>Decriptación (o descifrado).</b>
        <br>
        <br>
        El emisor puede recuperar el mensaje $m$ a partir de $c$ utilizando su llave privada $d$ de la siguiente forma: $$m \cong c^{d}\hspace{0.1cm}mod\hspace{0.1cm}n$$
        Teniendo $m$ entonces ya se podría recuperar el mensaje original $M$ invirtiendo el esquema que se acordó (también conocido como <i>padding scheme</i>).
</p>

    Siguiendo los pasos del numeral 1 procederemos a realizar la implementación del algoritmo:
    <br>
    <br>
    <b>Generación de las llaves</b>
    <br>
    <br>
    <ol>
    <li>A continuación se implementa el algoritmo de Fermat para determinar si dos números $p$ y $q$ generados aleatoriamente son primos. El proceso termina cuando tenemos dos números $p$ y $q$ primos los cuales se imprimen y se usarán para calcular $\phi(n)$ en el paso siguiente y los cuales para la implementación en la vida real se mantendrán en secreto. Se pondrá la condición que $p$ y $q$ estén en el intervalo $\left[256,10^{5}\right)$ esto para garantizar que $n$ sea mayor a 255 debido a que se utilizará el código ASCII de cada símbolo para poder convertir el mensaje $M$ en un número $m$ como se verá en el último paso de la implementación del algoritmo.</li>
    </ol>
</p>

```python
import random

def Primo_Fermat(x,k):
    if x%2==0:
        return 0
    for i in range(k):
        a = random.randint(2, x-1)
        if(not(((a**(x-1))%x)==1)):
            return 0
    return x

desicion = True
p=0;
q=0;
while(desicion):
    if (q==0):
        q = Primo_Fermat(random.randint(256, 10**5),50)
    elif (p==0):
        p = Primo_Fermat(random.randint(256, 10**5),50)
    elif (p==q):
        q =  Primo_Fermat(random.randint(2, 10),10)
    else:
        desicion = False
print("Primo p: ",p)
print("Primo q: ",q)
```

<p style='text-align: justify;'>
    <ol start="4">
    <li>Ahora calcularemos nuestra llave pública $e$ apoyándonos en el algoritmo de Euclides el cuál verificará que sea coprimo con $\phi(n)$ y que sea menor que este número. Se generará aleatoriamente y luego se procederá a la verificación.</li>
    </ol>
</p>

```python
import random

def MCD_Euclides(a,b):
    r0 = a
    r1 = b
    r = r0 % r1
    while(r!=0):
        r0 = r1
        r1 = r
        r = r0 % r1
    if(r1==1):
        return True
    return False
    
desicion = True
phi=2650443456
while(desicion):
    e = random.randint(2, phi)
    if((MCD_Euclides(phi,e))and(e<phi)):
        desicion = False
        
print("Exponente de encriptación e: ",e)
```

    </ol>
</p>

```python
def Inverso(a,b):
    aux = a
    s = 1
    t = 0
    sp = 0
    tp = 1
    while(b!=0):
        q = a // b
        r = a % b
        (a,s,t,b,sp,tp) = (b,sp,tp,r,s-sp*q,t-tp*q)
    if(t<0):
        t = t + aux
    return t

e = 473733239
phi = 2650443456
d = Inverso(phi,e)
print("Exponente de decriptación d: ",d)
```

<p style='text-align: justify;'>
    <b>Encriptación (o cifrado).</b>
    <br> 
    Para esta parte del cifrado y descifrado primero se presentará cada uno por separado introduciendo el número $m$ (<b>tal que $m<n$</b>) que representa el mensaje original $M$ y luego se presentará de manera completa las dos partes. Teniendo el entero $m$ que representa el mensaje se calculará $m^{e}\hspace{0.1cm} mod \hspace{0.1cm} n$ y como $m^{e}$ equivale a un número entre $0$ y $n-1$ en $\mathbf{Z}_{n}$ entonces podríamos escribir el mensaje cifrado como $c= m^{e} +n*k$ para algún número aleatorio $k\in \mathbf{Z}$ (sin pérdida de generalidad se establece que $0\leq k \leq n$ aunque podría ser cualquier entero).
</p>

```python
import random

def Pow(e,m,n):
    if(e==1):
        return m%n
    if(e%2==0):
        return ((Pow(e/2,m,n))**2)%n
    else:
        return (m*Pow(e-1,m,n))%n

m = 1548384
e = 473733239
n = 2650548587
c = Pow(e,m,n)+(n*random.randint(0, n))
print("Mensaje cifrado: ",c)
```

<p style='text-align: justify;'>
    <b>Decriptación (o descifrado).</b>
    <br> 
    
</p>

```python
def Pow(d,c,n):
    if(d==1):
        return c%n
    if(d%2==0):
        return ((Pow(d/2,c,n))**2)%n
    else:
        return (c*Pow(d-1,c,n))%n

d = 2556665927
c = 228402253056202278
n = 2650548587
m = Pow(d,c,n)
print("Mensaje descifrado m: ",m)
```

<p style='text-align: justify;'>
Para terminar, se presentará el cifrado y descifrado de manera completa y para esto se introducirá un mensaje $M$ el cual se convertirá en un arreglo de números cifrados mediante la utilización del código ASCII de cada símbolo.
</p>

```python
from numpy import array
def Pow(x,y,z):
    if(x==1):
        return y%z
    if(x%2==0):
        return ((Pow(x/2,y,z))**2)%z
    else:
        return (y*Pow(x-1,y,z))%z
    
def Encrypt(charA,e,n):
    ASCII = [0 for i in range(255)]
    m = [0 for i in range(len(charA))]
    print("Mensaje cifrado: ")
    for i in range (len(charA)):
        if(charA[i]==' '):
            m[i] = -1 #Los espacios en blancos se marcarán con el número -1.
            print(m[i]," - ", end="")
        else:
            if(ASCII[ord(charA[i])]!=0):
                m[i] = ASCII[ord(charA[i])]
                print(m[i]," - ", end="")
            else:
                m[i] = Pow(e,ord(charA[i]),n)
                print(m[i]," - ", end="")
    return m

def Decrypt(c,d,n):
    print()
    print("Mensaje descifrado: ")
    for i in range (len(c)):
        if (c[i]==-1):
            print(" ", end="")
        else:
            print(chr(Pow(d,c[i],n)), end="")
    
    
M = 'Matemásticas Discretas II, 2020-2'
charA = list(M)
e = 473733239
n = 2650548587
c = Encrypt(charA,e,n) # Se convierte al mensaje cifrado c en un solo paso.
d = 2556665927
Decrypt(c,d,n)
```
